```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import wordcloud
```

```python
df = pd.read_csv("2007/2007_22_keywords.csv", sep = '\t')
```

```python
df1 = df[(df['year'] <= 2012)]
df2 = df[(df['year'] >= 2013) & (df['year'] <= 2017)]
df3 = df[(df['year'] >= 2018)]
```

```python
#funkcia na vyrobenie word cloud z klucovych slov
def create_wordcloud(keywords):
    text = keywords
    word_cloud = WordCloud(
            background_color="white",  
            height=1000, 
            width=1200,
            ).generate(text)
    plt.imshow (word_cloud)
    plt.axis("off")
    plt.show()
```

```python
create_wordcloud(df1['tfidf_keywords'].str.cat())
```

```python
create_wordcloud(df2['tfidf_keywords'].str.cat())
```

```python
create_wordcloud(df3['tfidf_keywords'].str.cat())
```
